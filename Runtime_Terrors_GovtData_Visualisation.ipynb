```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV files
patient_data = pd.read_csv("/content/drive/MyDrive/HACKOUT/patient_data.csv")
hospital_data = pd.read_csv("/content/drive/MyDrive/HACKOUT/hospital_dataset.csv")

# Data Preprocessing and Analysis
# Hospital Beds and Facilities
total_icu_beds = hospital_data['icu_beds'].sum()
total_special_wards = hospital_data['special_wards'].sum()
total_semi_special_wards = hospital_data['semi_special_wards'].sum()
total_general_ward_beds = hospital_data['general_ward_beds'].sum()
total_oxygen_cylinders_capacity = hospital_data['oxygen_cylinders_capacity'].sum()

# Patient Data Analysis
top_findings = patient_data['finding'].value_counts().head(5)
average_age = patient_data['age'].mean()

# Visualization
# Hospital Beds and Facilities Visualization
labels = ['ICU Beds', 'Special Wards', 'Semi-Special Wards', 'General Ward Beds', 'Oxygen Cylinders Capacity']
sizes = [total_icu_beds, total_special_wards, total_semi_special_wards, total_general_ward_beds, total_oxygen_cylinders_capacity]
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Hospital Beds and Facilities Distribution')
plt.show()

# Top Findings Visualization
plt.figure(figsize=(10, 6))
plt.bar(top_findings.index, top_findings.values, color='b')
plt.title('Top 5 Findings among Patients')
plt.xlabel('Findings')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.show()

# Age Distribution Visualization
plt.figure(figsize=(8, 6))
plt.hist(patient_data['age'], bins=20, color='blue', edgecolor='black')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```
